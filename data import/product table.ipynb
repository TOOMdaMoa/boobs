```python
def parseItem(item):
    id = item.find("ITEM_ID").text
    url = item.find("URL").text
    img = item.find("MEDIA").find("IMGURL")
    if not img is None:
        img = img.text
    else:
        img = ""
    return {"id": id, "url": url, "img": img}

items = [parseItem(item) for item in root]
    
```
