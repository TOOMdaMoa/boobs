```python
cols_to_use = ['date',
               'variant_data_params_SIZE_EUR',
               'product_id',
               'user_id',
               'title_full_C',
               'variant_id',
               'variant_data_params_COLOR'
              ]
us = '0cf9fdec47251c5e8342e19c689192c5'
df2 = dff.loc[lambda x: x.user_id==us].sort_values('date')[cols_to_use].copy()
df2 = dff[cols_to_use].copy()
```

```python
def ff(x):
    x=x.sort_values('date')
    x['future_product']=x['product_id'].shift(-1)
    x['last_shoe']=x.product_id!=x.future_product
    k=[]
    i=0
    for j in x.last_shoe:
        k.append(i)
        if j:
            i=i+1
    x['single_product']=k
    y=x.groupby('single_product').apply(fff)
    return [y.count(),y.median(),y.var(),x['variant_data_params_SIZE_EUR'].median()]
def fff(x):
    x=x.sort_values('date')
    color= 'variant_data_params_COLOR'
    x['size_change']=(x['variant_id'].shift(1)!=x['variant_id'])&(x[color].shift(1)==x[color])
    to_return =list(x[x['size_change']]['variant_data_params_SIZE_EUR'])
    if (to_return!=[]): 
        return to_return[0]
    else:
        return np.nan
```

```python
g=df2.groupby('user_id')
df3=g.apply(ff)
```

```python
df5 = pd.DataFrame(df3.values.tolist(), columns = ['len','prediction','variance','median'])
df5['rozdil']=(df5.loc[:,'prediction']-df5.loc[:,'median'])
df5.index=df3.index
```
